```python
from datascience import Table, NumberFormatter
import datascience
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
%matplotlib inline

import numpy as np
```

```python
draft_data = [Table.read_table('data/drafts/draft_NBA_' + str(i) + '_stats.csv') if i > 1999 else None for i in range(0, 2016)]
anthro_data = [Table.read_table('data/drafts/draft_NBA_' + str(i) + '_stats.csv') if i > 1999 else None for i in range(0, 2016)]
```

```python
# removes NaNs in WS column for players who never made it to the NBA
for draft in draft_data:
    if draft != None:
        ws = draft.column('WS')
        ws[np.isnan(ws)] = np.nanmin(ws)
        draft['WS'] = ws
        
        pk = draft.column('Pk').astype(int)
        draft['Pk'] = pk

draft_data[2001].select(['Pk', 'WS']).show()
```

```python
win_shares = Table().with_column('Pk', np.arange(1, 59))

for i in range(0, len(draft_data)):
    draft = draft_data[i]
    if draft != None:
        ws = draft.column('WS')

        while len(ws) < 58:
            ws = np.append(ws, 0)
        
        if len(ws) > 58:
            ws = ws[:58]
            
        total = np.sum(ws)
        
        win_shares['WS' + str(i)] = ws / total

win_shares['WSavg'] = np.array([np.mean(win_shares.row(i)[1:]) for i in range(0, win_shares.num_rows)])
win_shares.show()
win_shares.select(['Pk', 'WSavg']).scatter('Pk', overlay=True, fit_line=True)

# plt.plot(win_shares.column('WS2000'))
```
